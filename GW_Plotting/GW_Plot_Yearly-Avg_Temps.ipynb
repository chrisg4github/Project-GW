## What is a lag plot used for?
A lag plot is a scatter plot for a time series and the same data lagged.  We are frequently interested in whether or not data which we have collected is random. Lag plots are used to investigate the randomness of a dataset. If the data is in fact random, then the lag plot will exhibit no structure, while nonrandom data will exhibit some kind of structure. Unfortunately, this does not give us an idea of
what exactly that structure may be, but it is a quick and eective way to investigate the randomness of a given dataset.

```python
from pandas.plotting import lag_plot

plt.figure()
data = gltc1_df["yr_avg_temp_f"]
lag_plot(data)
plt.title("USA 1878-2012 Yearly Avg TmpF Lag Plot")
plt.tight_layout()
plt.savefig("./images/usa_lag-plot_temp.png")
plt.show()
```

## What is a histogram used for?
Histograms show the frequency and distribution of a data column or feature.  The y values are the frequency and x values are the distribution values put into bins.  Bin are set with a constant like bins=20 or bins='auto' which lets matplotlib choose the number of bins.

```python
from matplotlib import pyplot as plt 
import numpy as np  
   
a = np.array(gltc1_df["yr_avg_temp_f"]) 
plt.hist(a, bins='auto') 
plt.title("USA 1878-2012 Yearly Avg TmpF Histogram")
plt.tight_layout()
plt.savefig("./images/usa_yly_avg_tmpf_histogram.png") 
plt.show()
```

```python
# Dependencies
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime, date, time
import matplotlib.dates as mdates

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

x_axis = gltc1_df["year"]  
y1_axis = gltc1_df["yr_avg_temp_f"]
y8_axis = gltc8_df["yr_avg_temp_f"]
y9_axis = gltc9_df["yr_avg_temp_f"]




ax.set_title("Yearly Avg Temp(F) USA-AU-IN",\
             weight='bold').set_fontsize('18')
ax.set_xlabel("1878 - 2012", weight='bold').set_fontsize('14')
ax.set_ylabel("Yearly Avg Temps(F) USA-AU-IN", 
              weight='bold').set_fontsize('14')

ax.grid(True)
ax.set_facecolor("whitesmoke")
fig.autofmt_xdate(rotation=45, ha='right')

ax.plot(x_axis, y1_axis, color="magenta", linewidth= 2)
ax.plot(x_axis, y8_axis, color="limegreen", linewidth= 2)
ax.plot(x_axis, y9_axis, color="royalblue", linewidth= 2)



ax.legend(["USA Avg(F)","Australia Avg(F)","India Avg(F)"],
          bbox_to_anchor=(1.0,.65))

plt.tight_layout()
plt.savefig("./images/yearly_avg_tempf_usa-au-in.png")
plt.show()
```
