# The story of CO2 emissions:

Attention is today focused on exactly when human activities started to significantly influence global warming.  Due to current data tracking, it is widely accepted that human influence became significant around the beginning of the industrial age in the 1850â€™s.  A higher level of CO2 emissions is observed beginning from about 1960. 

When fossil fuels are burned, carbon dioxide is added to the atmosphere; most of it cannot be absorbed into the natural carbon cycle.  The emissions remaining in the atmosphere absorb heat and thus contribute to the gradual temperature increase across the globe.


```python
import pandas as pd
import numpy as np
import scipy as sp
import sklearn as sk
import matplotlib.pyplot as plt
# import the date module
from datetime import date
```

```python
# Store filepath in a variable: 
#co2_1 = "co2.csv"
co2_1 = "CO2_bynation_1751_2014.csv"
co2_1_df = pd.read_csv(co2_1)
#co2_1_df = pd.read_csv(co2_1,parse_dates=['year'])
co2_1_df.info()
co2_1_df.head()
```

```python
# Rename the data columns
co2_r_df=co2_1_df.rename(columns={'Nation':'country',
                             'year': 'year', 
                              'co2' :'co2' })
co2_r_df.head()
```

```python
# drop the NaN values and note the details of the dropped data.
co2_r_df= co2_r_df.dropna(axis=0)
co2_r_df.describe()
co2_r_df.info()
```

```python
#determine   what  is the date  range of df  at this step

co21_min_date = co2_r_df["year"].min()
co21_max_date = co2_r_df["year"].max()
print("min date " + str(co21_min_date) + " max date " + str(co21_max_date))
```

```python
# Export file as an XLSX or CSV, w/o index, w/ header

#co2_r_df.to_excel("./clean_GlobalCO2EmissionsByCountry.xlsx", index=False, header=True)
co2_r_df.to_csv("./clean_bynationCO2EmissionsByCountry.csv", index=False, header=True)
```
